# Tutorial 27 : Pandas : iloc(we will use this instead of square brackets..)

```python
movies_csv.head(10)
```

```python
# Select single column
movies_csv['original_title']  # Panda series...
```

```python
# Select multiple columns..
movies_csv[['original_title', 'budget']]
```

```python
movies_csv[0]  # Select row.. it will give an error...
```

```python
movies_csv[0 : 1]  # Slicing is used here for accessing the rows
```

```python
# We select multiple rows like this...
movies_csv[5 : 15]
```

```python
# Here, we will use iloc for selecting rows and columns...
```

```python
movies_csv.iloc[0]  # this prints the 0th row values...
```

```python
type(movies_csv.iloc[0])  # here, we get the type as Pandas ki Series..
```

```python
movies_csv.iloc[0 : 5]
```

```python
type(movies_csv.iloc[0 : 5])
```

```python
movies_csv.iloc[[0, 5, 10, 15]]  # Here, we have passed the list of row index for those of which we need the data...
```

```python
movies_csv.iloc[[0, 5, 10, 15], [6, 1]]  # movies_csv.iloc[row ki selection k liye , columns ki selection k liye]
```

```python
movies_csv.iloc[[0, 5, 10, 15], 0 : 7] # what if need from 'budget' to till 'original_title', here we use slicing i.e 0:7
```

```python
movies_csv.iloc[[0, 5, 10, 15], 5 : 15]
```

```python
# If we want to target rows like, 0,5,10, 15 and columns from 0th to till 4th then we do this...
movies_csv.iloc[[0,5,10,15], : 5]
```
