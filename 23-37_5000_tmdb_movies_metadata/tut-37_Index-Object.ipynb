### Reset Inex, Set Index, Indexing and Slicing of index object, get_loc, is_unique, create index object from scratch

```python
import pandas as pd
```

```python
movies_df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies_df.head() # This will return the movies csv file's top 5 rows..
```

```python
# In order to use the 'original_title' as index column:-
movies_df1 = pd.read_csv('tmdb_5000_movies.csv', index_col = 'original_title')
```

```python
movies_df1
```

```python
# By this way we can get the details of the 'Spectre' using it as index :-
movies_df1.loc['Spectre']
```

```python
# It will reset the index with the previous ones: function used : reset_index()
# But this will not make changes to the original data, it will just show it...
movies_df1.reset_index()
```

```python
movies_df1.head()  # we can see that original data has not been changed..
```

```python
# Now, to make changes to the original data as well, we do this:
movies_df1.reset_index(inplace = True)
```

```python
movies_df1.head() # Now it got changed to original data
```

```python
# Now, we want to make the column as index of Original_title and make this change to original data as well:
movies_df1.set_index('original_title', inplace = True)
```

```python
movies_df1.head()
```

```python
movies_df1.index  # by this we will get all the index in our dataFrame.
```
