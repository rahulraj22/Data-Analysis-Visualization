```python
# What if just want the data of Israel of year- 2014 ?
mask1 = df.country == 'Ukraine'
mask2 = df.year == 2014
mask3 = df.year == 2015
f_df = df[mask1 & (mask2 | mask3)]  # This will give the results Of 'Israel' but of year 2014 and 2015
```

```python
f_df
```

```python
# what if we don't need this much columns ?
# here we will loc operator:
f_df1 = df.loc[mask1 & (mask2 | mask3), ['country', 'year', 'sex', 'age', 'suicides_no']]
```

```python
f_df1
```

```python
# Here, we see that males and females both coming our dataFrame, what if we need only male data of Israel:
# then , here we will add one more mask:
mask4 = df.sex == 'male'
f_df2 = df.loc[mask1 & mask4 & (mask2 | mask3), ['country', 'year', 'sex', 'age', 'suicides_no']]
```

```python
f_df2
```

```python
# Lets find the sum of all suicides in above dataFrame i.e f_df2:
# Here, 563 is the total no. of suicides happened in 2014 and 2015 by males of Israel
f_df2.sum().suicides_no
```

```python
# Lets see for females suicides data:--
mask5 = df.sex == 'female'
f_df3 = df.loc[mask1 & mask5 & (mask2 | mask3), ['country', 'year', 'sex', 'age', 'suicides_no']]
f_df3
```

```python
f_df3.sum().suicides_no
```
