```python
import pandas as pd
```

```python
df = pd.read_csv('suicide_dataset.csv')
```

```python
df
```

```python
# Here, we can see that there is data begining from 1987 and so on but we need data from 2010, so to do ?
mask1 = df.year >= 2010
df.loc[mask1]
```

```python
# What if we need data between two interval of year, lets create the mask for it:
mask2 = df.year.between(2000, 2010)
df.loc[mask2]
```

```python
# What if we just need data of two year ?
mask3 = df.year.isin([2000, 2010])
df.loc[mask3]
```

```python
# What if specifically don't need data of 2000, 2010 year:
f_df = df.loc[~mask3]
```

```python
f_df
```

```python
f_df.year.unique()
```

### Now, lets do an Exercise: I need a data of suicides b/w year 2010 and 2014 males and females ?

```python
# 2010 - 2014:
# Lets see of place: Switzerland: males | females | suicide data:
Mask1 = df.year.between(2010, 2014)
Mask2 = df.sex == 'female'
Mask3 = df.country == 'Switzerland'
Mask4 = df.age == '15-24 years'
f_df_Switzerland = df.loc[Mask1 & Mask2 & Mask3 & Mask4, ['country', 'year', 'sex', 'age', 'suicides_no']]
# ['country', 'year', 'sex', 'age', 'suicides_no'] implies that this much columns we need in our dataFrame.
```
